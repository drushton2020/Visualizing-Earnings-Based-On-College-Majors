## Visualizing Earnings Based On College Majors

We'll be using data found in [five-thirtyeight's github repo](https://github.com/fivethirtyeight/data/tree/master/college-majors) on the earnings and employment statistics of different college majors between 2010 and 2012. The [raw data](https://www.census.gov/programs-surveys/acs/) was originally published by the American Community Survey.

## Exploring and Cleaning

```python
import pandas as pd
import matplotlib as plt
```

```python
%matplotlib inline
```

```python
recent_grads = pd.read_csv("recent-grads.csv")
```

```python
recent_grads.iloc[0]
```

```python
recent_grads.head()
```

```python
recent_grads.tail()
```

```python
raw_data_count = recent_grads.shape
print(raw_data_count)
```

```python
recent_grads = recent_grads.dropna()
```

```python
cleaned_data_count = recent_grads.shape
print(cleaned_data_count)
```

## Scatter Plots

```python
recent_grads.plot(
                    x='Sample_size', 
                    y='Median', 
                    kind='scatter',
                    title='Sample Size vs Median')
```

```python
recent_grads.plot(
                    x='Sample_size', 
                    y='Unemployment_rate', 
                    kind='scatter',
                    title= 'Sample Size vs Unemployment Rate')
```

```python
recent_grads.plot(
                    x='Full_time', 
                    y='Median', 
                    kind='scatter',
                    title='Full Time vs Median')
```

```python
recent_grads.plot(
                    x='ShareWomen', 
                    y='Unemployment_rate', 
                    kind='scatter',
                    title='ShareWomen vs Unemployment Rate')
```

```python
recent_grads.plot(
                    x='Men', 
                    y='Median', 
                    kind='scatter',
                    title='Men vs Median')
```

```python
recent_grads.plot(
                    x='Women', 
                    y='Median', 
                    kind='scatter',
                    title='Women vs Median')
```
